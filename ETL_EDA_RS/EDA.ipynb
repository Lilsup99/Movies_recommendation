Parece estar relativamente completo el dataset, la grande cantidad de nulos en 
`belongs_to_collection` es porque no todas las peliculas forman parte de una serie o coleccion de peliculas

```python
df.loc[df['title'].isna()]
```

estos registros parecen no aportar nada, los elimino

```python
df = df.drop(12718, axis=0)
df = df.drop(12882, axis=0)
df = df.drop(12905, axis=0)
```

chequeamos que no hayan registros duplicados

```python
df.loc[df['title'].duplicated()].dropna()
```

hay algunas peliculas repetidas, las voy a eliminar

```python
df.query('title == "Hamlet"')
```

Observo que hay registros con titulos de peliculas duplicados, si bien no son la misma pelicula, son peliculas que tratan de lo mismo, para efecto practico del sistema de recomendacion, eliminare los registros duplicados de la columna `title`

```python
df = df.loc[~df['title'].duplicated()]
```
