# Desanidar las columnas

#### `belongs_to_collection`

```python
eval(data['belongs_to_collection'][0])
```

en cada registro de la columna `belongs_to_collection`, se reemplazara, en caso de no ser nulo, el valor del dict con el nombre

```python
for i in range(len(data)):
    try:
        if isinstance(eval(data['belongs_to_collection'][i]), dict):
            data['belongs_to_collection'][i] = eval(data['belongs_to_collection'][i])['name']
    except:
        pass        
```

#### `genres`

creare una columnas en el dataframe donde se almacenaran los generos en una tupla.

```python
pd.json_normalize(eval(data['genres'][5]))
```

```python
## se desanida la columna genres en 2 columnas
data['Genres'] = np.nan




for i in range(len(data)):
    try:
        dft = pd.json_normalize(eval(data['genres'][i]))
        gen = []
        for j in range(len(dft)):
            gen.append(dft['name'][j])
        data['Genres'][i] = gen
    except: continue       
    
```

```python
data
```

#### `production_companies`

se procedera similar como se hizo en la columna de genres

```python
#desanido la columna 'production_companies'

data['pro_comp1'] = np.nan
data['pro_comp2'] = np.nan

for i in range(len(data)):
    try:
        dft = pd.json_normalize(eval(data['production_companies'][i]))

        for j in range(len(dft)):
            data[f'pro_comp{j+1}'][i] = dft['name'][j]
    except:
        continue 
               
            
```

#### 'production_countries'

se crea una columna nueva y se almacena el pais por cada pelicula

```python
data['production_countries'][489]
```

```python
# Pais
data['country'] = np.nan

for i in range(len(data)):
    try:
        dft = pd.json_normalize(eval(data['production_countries'][i]))

        for j in range(len(dft)):
            data['country'][i] = dft['name'][j]
    except:
        continue   
```

#### `spoken_languages`

```python
data['language'] = np.nan

for i in range(len(data)):
    try:
        dft = pd.json_normalize(eval(data['spoken_languages'][i]))

        for j in range(len(dft)):
            data['language'][i] = dft['name'][j]
    except:
        continue 
```

```python
data
```

```python
'''
se eliminan las columnas que desanide
'''

columnas = ['genres', 'production_companies',
       'production_countries','spoken_languages']

data = data.drop(columns=columnas)
```

```python
data
```

Para la ultima funcion de la API se necesitan los datos de los directores de las peliculas, en este paso los sacare del dataset `credits.csv` y los incorporare a `data`

```python
df = pd.read_csv('PI MLOps\credits.csv')
```

```python
pd.DataFrame(eval(df['crew'][45441]))
```

```python
x = pd.DataFrame(eval(df['crew'][4000]))
(x['name'][x['job'] == 'Director'].index)[0]
```

```python
df['director'] = np.nan
for i in range(len(df)):
    try:
        x = pd.json_normalize(eval(df['crew'][i]))
        df['director'][i] = x['name'][x['job'] == 'Director'][(x['name'][x['job'] == 'Director'].index)[0]]
    except:
        continue
```

ahora con los actores

```python
pd.DataFrame(eval(df['cast'][1000]))['name'][0]
```

Me quedare con los primeros 3 actores de cada pelicula

```python
df['actors'] = np.nan

for i in range(len(df)):
    act = []
    x = pd.DataFrame(eval(df['cast'][i]))
    try:
        for j in range(3):
            act.append(x['name'][j])
        df['actors'][i] = act
    except: continue        
```
